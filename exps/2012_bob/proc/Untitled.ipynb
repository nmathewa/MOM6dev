```python
import os
import xarray as xr
os.chdir("/home/nma/mom/MOM6dev/exps/2012_bob/proc/")
```

```python
era_data = xr.open_dataset("2012-2013_era5daily.nc")

u10,v10 = era_data.u10,era_data.v10

vel_file = u10.to_dataset()

vel_file["v10"] = v10

vel_file.to_netcdf("era_uv.nc")
```

```python
t2m = era_data.t2m.to_dataset()

ct2m = t2m.t2m

fint2m = ct2m.to_dataset()

fint2m.to_netcdf("era_t2m.nc")
```

```python
finslp = era_data.msl.to_dataset()

finslp.to_netcdf("era_slp.nc")
```

```python
sw,lw = era_data.msdwswrf,era_data.msdwlwrf


finflux = sw.to_dataset()

finflux["lw"] = lw

finflux.to_netcdf("era_flux.nc")
```
