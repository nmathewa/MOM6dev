```python
n_lon,n_lat = lon[::2],lat[::2]

n_lonss = []
n_latss = []
for ii in range(len(n_lon)):
    n_lonss += [(n_lon[ii-1] + n_lon[ii])/2]

for jj in range(len(n_lat)):
    n_latss += [(n_lat[jj-1] + n_lat[jj])/2]

new_lons,new_lats = n_lonss[1:],n_latss[1:]
```
